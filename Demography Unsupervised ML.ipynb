```python
demo=data.values
demo
```

```python
from sklearn.preprocessing import StandardScaler,normalize
std=StandardScaler()
stand=std.fit_transform(demo)
norm=normalize(stand)
print('StandardScaler',stand)
print('Normalize',norm)
```

```python
from sklearn.decomposition import PCA
pca=PCA(2)
a=pca.fit(norm)
print(a)
print(pca.components_)
B=pca.transform(norm)
print(B)
```

```python
error=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i).fit(B)
    error.append(kmeans.inertia_)
print(error)
```

```python
plt.title('Elbow Method')
plt.plot(range(1,11),error)
plt.xlabel('No. of clusters')
plt.ylabel('Error')
plt.show()
```
